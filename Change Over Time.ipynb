```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('homeowners_over_time.csv')

print(df.columns)
```

```python
# Convert 'year' column to numeric type
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['change'], marker='o', linestyle='-')
plt.title('Change Over Years')
plt.xlabel('Year')
plt.ylabel('Change')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
